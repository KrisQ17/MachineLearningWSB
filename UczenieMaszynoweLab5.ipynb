```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/palles77/MachineLearning_Exercise4/main/Employee_Compensation_SF.csv')
```

```python
# function returning column index in a dataframe
def column_index(df, query_cols):
    cols = df.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols,query_cols,sorter=sidx)]
```

```python
# reading 'Total Benefits' -> column 20, 'Total Compensation' -> column 21
(column_first, colum_second) = column_index(dataset, ['Total Benefits', 'Total Compensation'])
# get sample 1000 rows
dataset_for_clustering = dataset.sample(1000).iloc[:, [column_first, colum_second]].values
```

```python
# Drawing dendrogram
dendrogram = sch.dendrogram(sch.linkage(dataset_for_clustering, method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters = 3, affinity='euclidean', linkage='ward')
model.fit(dataset_for_clustering)
labels = model.labels_
```

```python
# Drawing Clusters
plt.scatter(dataset_for_clustering[labels==0, 0], dataset_for_clustering[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(dataset_for_clustering[labels==1, 0], dataset_for_clustering[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(dataset_for_clustering[labels==2, 0], dataset_for_clustering[labels==2, 1], s=50, marker='o', color='green')
#plt.scatter(dataset_for_clustering[labels==3, 0], dataset_for_clustering[labels==3, 1], s=50, marker='o', color='purple')
#plt.scatter(dataset_for_clustering[labels==4, 0], dataset_for_clustering[labels==4, 1], s=50, marker='o', color='orange')
plt.xlabel('Total Benefits')
plt.ylabel('Total Compensation')
plt.show()
```

```python
# Zadanie 9, 10, 11, 12, 13, 14
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
features, true_labels = make_blobs(
    n_samples=1000,
     centers=5,
    cluster_std=2.5,
    random_state=42
)
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features[:5]
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=5,
    n_init=10,
    max_iter=300,
    random_state=42)
kmeans.fit(scaled_features)
```

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
    
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(
    range(1, 11), sse, curve="convex", direction="decreasing"
)

kl.elbow
silhouette_coefficients = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
    
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.metrics import adjusted_rand_score
```

```python
features, true_labels = make_moons(
    n_samples=250, noise=0.05, random_state=42
)
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(n_clusters=2)
dbscan = DBSCAN(eps=0.3)
kmeans.fit(scaled_features)
dbscan.fit(scaled_features)
kmeans_silhouette = silhouette_score(
    scaled_features, kmeans.labels_
).round(2)
dbscan_silhouette = silhouette_score(
   scaled_features, dbscan.labels_
).round (2)
```

```python

fig, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(8, 6), sharex=True, sharey=True
)
fig.suptitle(f"Clustering Algorithm Comparison: Crescents", fontsize=16)
fte_colors = {
    0: "#008fd5",
    1: "#fc4f30",
}

km_colors = [fte_colors[label] for label in kmeans.labels_]
ax1.scatter(scaled_features[:, 0], scaled_features[:, 1], c=km_colors)
ax1.set_title(
    f"k-means\nSilhouette: {kmeans_silhouette}", fontdict={"fontsize": 12}
)


db_colors = [fte_colors[label] for label in dbscan.labels_]
ax2.scatter(scaled_features[:, 0], scaled_features[:, 1], c=db_colors)
ax2.set_title(
    f"DBSCAN\nSilhouette: {dbscan_silhouette}", fontdict={"fontsize": 12}
)
plt.show()
```

Klasterowanie to technika nienadzorowanego uczenia maszynowego, grupująca punkty danych w klastry na podstawie podobieństwa informacji dostępnych punktów w zestawie danych.  
- K-means i DBScan to dwa najpopularniejsze algorytymy do klastrowania.  
- W K-means skupiska mają kulisty bądź wypukły kształt i muszą mieć tę samą wielkość cechy, natomiast w DBScan tworzone klastry mają dowolny kształt i nie muszą mieć tej samej wielkości cechy.  
- K-means nie działa dobrze z wartościami granicznymi i dużym szumem danych, natomiast DBScan jest bardziej uniwersalne w tym temacie.  
- W DBScan nie trzeba określać liczy klastrów do klasteryzacji, natomiast K-means jest wrażliwe na liczbę skupień.  
- K-means jest bardziej wydajne w przypadku dużych zestawów danych, a DBScan może wtedy nie być taki wydajny.
